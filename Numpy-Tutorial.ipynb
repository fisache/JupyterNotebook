# Broadcast

<pre>
다른 차원의 배열끼리 연산을 수행할 경우,
배열을 같은 형상으로 만든 후 연산을 진행한다.
ex) 2*2 + 1*2 ==> 1*2를 2*2로 변형한다.
[1, 2]    +  [1, 2] ==> [1, 2]   +  [1, 2]
,[3, 4]                 ,[3, 4]     ,[1, 2] # 위 행의 [1,2]를 따라간다.
</pre>

```python
A = np.array([[1,2],[3,4]])
```

```python
B = np.array([[10,20]])
```

```python
A * B # Broadcast
```

# Access to index

```python
x = np.array([[51, 55], [14, 19], [0 ,4]])
```

```python
x
```

```python
x[0][1] # 0 Row, 1 Col
```

```python
for row in x:
    print(row)
```

```python
for row in range(0, len(x)):
    for col in range(0, len(x[0])):
        print("row: " , row, " col: ", col, " value: ", x[row][col])
```

```python
x = x.flatten() # N-dimesion -> 1-dimesion
```

```python
x
```

<pre>
일반적인 반복(for)과 filter(v>15)을 간단히 표현할 수 있다.
</pre>

```python
_x = []
for v in x:
    if v > 15: _x.append(v)
_x
```

```python
x > 15 # filtering
```

```python
_x = x[x>15] # apply to filtering
_x
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) # 0부터 6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin') # plot 이름
plt.plot(x, y2, linestyle='--', label='cos') # plot 이름, 라인 점선
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('sin & cos') # 그래프 이름
plt.legend()
plt.show()
```
